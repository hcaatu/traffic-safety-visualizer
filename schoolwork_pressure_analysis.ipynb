```python
df = pd.read_csv("school.csv")

hours_df = df[[
    "Gender",
    "Ageyears",
    "Schoolwork_Pressure",
    "Score_in_memory_game",
    "Sleep_Hours_Schoolnight", 
    "Sleep_Hours_Non_Schoolnight", 
    "Hanging_Out_With_Friends_Hours",
    "Talking_On_Phone_Hours",
    "Doing_Homework_Hours",
    "Doing_Things_With_Family_Hours",
    "Outdoor_Activities_Hours",
    "Video_Games_Hours",
    "Social_Websites_Hours",
    "Texting_Messaging_Hours",
    "Computer_Use_Hours",
    "Watching_TV_Hours",
    "Paid_Work_Hours",
    "Work_At_Home_Hours"
]]

for col in hours_df.columns:
    if col in ["Gender", "Schoolwork_Pressure"]:
        continue
    hours_df[col] = pd.to_numeric(hours_df[col], errors='coerce')

data = hours_df.groupby("Schoolwork_Pressure").mean(numeric_only=True).reset_index()

data = data.drop(["Ageyears", "Score_in_memory_game", "Computer_Use_Hours", "Sleep_Hours_Schoolnight", "Sleep_Hours_Non_Schoolnight"], axis=1)
data = data.rename(columns=lambda x: x.replace('_', ' '))
data = data.rename(columns=lambda x: x.replace(' Hours', ''))

reshaped_df = data.melt(
    id_vars="Schoolwork Pressure",
    var_name="variable",
    value_name="average"
)

reshaped_df = reshaped_df.rename(columns=lambda x: x.replace('_', ' '))

fig = px.bar(
    reshaped_df,
    x="variable",
    y="average",
    color="Schoolwork Pressure",
    color_discrete_sequence=['#185234', '#249c5e', '#89d94c'],
    barmode="group",
    title="",
    labels={"average": "Hours a Week", "variable": "", "Schoolwork Pressure": "Experienced Pressure"}
)

fig.show()
```
